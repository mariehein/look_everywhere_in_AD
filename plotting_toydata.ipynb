```python
import plotting_utils as pl
import numpy as np
```

```python
general_directory = "/hpcwork/zu992399/look_elsewhere_paper/"
split_into =10
```

# Binned classifier

```python
direc = general_directory+"toydata_binned/"
NNBDT =False
p_train = np.sort(np.load(direc+"p_train.npy"))
p_test = np.sort(np.load(direc+"p_test.npy"))
p_kfolds = np.sort(np.load(direc+"p_kfolds.npy"))
p_train = pl.prep_pvalues(p_train, split_into, NNBDT)
p_test = pl.prep_pvalues(p_test, split_into, NNBDT)
p_kfolds = pl.prep_pvalues(p_kfolds, split_into, NNBDT)
pl.calibration_curve_plot_errorband(p_train, p_test, p_kfolds, 100000, "toydata_binned", title="Binned", NNBDT=NNBDT)
```
