```python
import plotting_utils as pl
import numpy as np
import statistics_utils as stats
```

```python
general_directory = "/hpcwork/zu992399/look_elsewhere_paper/"
split_into =10
two_sided = False
```

# Binned classifier

```python
direc = general_directory+"toydata_binned_R/"
NNBDT =False
if two_sided:
    p_train = np.sort(np.load(direc+"p_train.npy"))
    p_test = np.sort(np.load(direc+"p_test.npy"))
    p_kfolds = np.sort(np.load(direc+"p_kfolds.npy"))
else:
    p_train = np.sort(np.load(direc+"p_train_onesided.npy"))
    p_test = np.sort(np.load(direc+"p_test_onesided.npy"))
    p_kfolds = np.sort(np.load(direc+"p_kfolds_onesided.npy"))
p_train = pl.prep_pvalues(p_train, split_into, NNBDT)
p_test = pl.prep_pvalues(p_test, split_into, NNBDT)
p_kfolds = pl.prep_pvalues(p_kfolds, split_into, NNBDT)
pl.calibration_curve_plot_errorband(p_train, p_test, p_kfolds, 100000, "toydata_binned_R", NNBDT=NNBDT, title="Binned")
```

# NN total

```python
direc = general_directory+"toydata_NN_noearlystopping/"
NNBDT=True
p_train = np.sort(stats.p_values_from_folder(direc+"traintest/", datatype="train", two_sided=two_sided))
p_test = np.sort(stats.p_values_from_folder(direc+"traintest/", datatype="test", two_sided=two_sided))
p_kfolds = np.sort(stats.p_values_from_folder(direc+"kfolds/", datatype="kfolds", two_sided=two_sided))
p_train = pl.prep_pvalues(p_train, split_into, NNBDT)
p_test = pl.prep_pvalues(p_test, split_into, NNBDT)
p_kfolds = pl.prep_pvalues(p_kfolds, split_into, NNBDT)

direc = general_directory+"toydata_NN/"
p_train2 = np.sort(stats.p_values_from_folder(direc+"traintest/", datatype="train", two_sided=two_sided))
p_test2 = np.sort(stats.p_values_from_folder(direc+"traintest/", datatype="test", two_sided=two_sided))
p_kfolds2 = np.sort(stats.p_values_from_folder(direc+"kfolds/", datatype="kfolds", two_sided=two_sided))
p_train2 = pl.prep_pvalues(p_train2, split_into, NNBDT)
p_test2 = pl.prep_pvalues(p_test2, split_into, NNBDT)
p_kfolds2 = pl.prep_pvalues(p_kfolds2, split_into, NNBDT)
pl.calibration_curve_plot_errorband_multiple([p_train, p_train2], [p_test, p_test2], [p_kfolds, p_kfolds2], 10000, "toydata_NN_total", NNBDT=True, title="NN")
```
